```python
from requests import get
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vSc_2y5N0I67wDU38DjDh35IZSIS30rQf7_NYZhtYYGU1jJYT6_kDx4YpF-qw0LSlGsBYP8pqM_a1Pd/pubhtml#'
response = get(url)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
list_data = []
for a in soup.findAll('table'):
    for i in a.findAll('tr'):
        data = []
        for b in i.findAll('td'):
            #print(b.text)
            data.append(b.text)
        list_data.append(data)

print(list_data)
```

```python
list_data = list_data[1:1061]
```
